# Simplified view of snapshots

The [snapshot-view-demo notebook](snapshot-view-demo.ipynb) had a lot of detail.  Let's look at a more concise scenario.

```python
from databroker import Broker
from snapshot import SnapshotReport

# standard local configuration using mongodb
db = Broker.named("mongodb_config")

# our custom callback
snapshot_report = SnapshotReport()
```

```python
# pick the first snapshot the databroker finds
h = list(db(plan_name="snapshot", since="2018-12-19", until="2018-12-19 23:59:59"))[0]
snapshot_report.print_report(h)
```

or to view all the snapshots from Dec 19:

```python
for h in db(plan_name="snapshot", since="2018-12-19", until="2018-12-19 23:59"):
    snapshot_report.print_report(h)
```
